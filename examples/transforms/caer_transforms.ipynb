```python
# Imports
import caer 
import caer.transforms as ct
import matplotlib.pyplot as plt 

def imshow(tens):
    # We expect `tens` to be an RGB image (default in Caer)
    plt.imshow(tens)
    plt.xticks([]), plt.yticks([])  # Hides the graph ticks and x / y axis
    plt.margins(0,0)
    plt.show()

# Creating our image
tens = caer.imread(r'https://images.unsplash.com/photo-1582949663569-4ae9faa3d4b8?ixid=MXwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHw%3D&ixlib=rb-1.2.1&auto=format&fit=crop&w=755&q=80', rgb=True)
imshow(tens)
```

Augmentations against Caer

1. Flips

a. Horizontal Flip (``caer.transforms.hflip()``)

```python
hf = ct.hflip(tens)
imshow(hf)
```

b. Vertical Flip (``caer.transforms.vflip()``)

```python
vf = ct.vflip(tens)
imshow(vf)
```

c. Horizontal & Vertical Flip (``caer.transforms.hvflip()``)

```python
hvf = ct.hvflip(tens)
imshow(hvf)
```

d. Random Flip (``caer.transforms.rand_flip()``)

```python
randf = ct.rand_flip(tens)
imshow(randf)
```

2. Rotations

```python
# Rotate around the centre (``rotPoint=None``)
rotate = ct.rotate(tens, 45, rotPoint=None)
imshow(rotate)
```

```python
# Rotate around (100,100)
rotate100 = ct.rotate(tens, 45, (100,100))
imshow(rotate100)
```
