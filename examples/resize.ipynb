# Resizing Images

With `caer.resize()`, you can resize Caer-processed images to various sizes with the following functionalities:

1. Using a resize factor (keeps original aspect ratio)
2. Using a fixed target size (optional: keep the original aspect ratio)

```python
import caer 

# We need Matplotlib to display the images
import matplotlib.pyplot as plt 
```

```python
# Loading a standard image that Caer ships with
img = caer.data.sunrise(rgb=True)

# Image Size
# Note that img.shape might follow the (height, width) format as Caer internally uses Numpy to process images
print(img.shape)

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Currently, our image has a size (640,427)
# To resize the image to (230,200) --> follows the (width,height) convention
resized = caer.resize(img, target_size=(230,200))

# Resize keeping the original aspect ratio
resized_wratio = caer.resize(img, target_size=(230,200), keep_aspect_ratio=True)

# Resize using a resize_factoer --> Scales the image by .2
resized_wfactor = caer.resize(img, resize_factor=.2)

# Display image sizes
print(f'Original Image size = {img.shape}')
print(f'Hard Resized Image size = {resized.shape}')
print(f'Aspect Ratio Image size = {resized_wratio.shape}')
print(f'Resize Factor Image size = {resized_wfactor.shape}')

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(20,7))

ax1.imshow(img)
ax1.set_title('Original')
ax1.yaxis.tick_right()

ax2.imshow(resized)
ax2.set_title('Hard Resize')
ax2.yaxis.tick_right()

ax3.imshow(resized)
ax3.set_title('Aspect Ratio')
ax3.yaxis.tick_right()

ax4.imshow(resized_wfactor)
ax4.set_title('Resize Factor')
ax4.yaxis.tick_right()
```
